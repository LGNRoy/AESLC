```python
subcount = {}
count = 0
for line in open("random_raw/full.txt"):
    subtitle = line.split("<sep>")[1]
    if subtitle not in subcount:
        subcount[subtitle] = 0
    subcount[subtitle] += 1
    count += 1
```
